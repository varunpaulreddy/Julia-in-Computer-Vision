```julia
using Pkg
Pkg.add("PyCall")
Pkg.add("ImageView")
Pkg.add("Images")
```

```julia
using PyCall

# Import OpenCV via PyCall
const cv2 = pyimport("cv2")

# Load an image
image_filepath = "people.jpg"
img = cv2.imread(image_filepath)

# Create a window to display the image
window_name = "Image Window"
cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)
cv2.imshow(window_name, img)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


#faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

window_name = "Gray Window"
cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)
cv2.imshow(window_name, gray)

# Detect faces in the grayscale frame using haar
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)


# Wait for a key press and close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```julia
# Import OpenCV via PyCall
using Images, ImageDraw

img = load("people.jpg")

for face in eachrow(faces)
    x, y, w, h = face
    println(x, " ", y, " ", w, " ", h)
    #img = draw!(img, Polygon(RectanglePoints(Point(10, 10), Point(100, 100))))
    img = draw!(img, Polygon(RectanglePoints(Point(x, y), Point(x+w, y+h))), RGB{N0f8}(1))
end
```
